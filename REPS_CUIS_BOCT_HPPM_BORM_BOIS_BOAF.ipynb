```python
debt_disbursement=boct_non_zero[boct_non_zero['DESCRIPTION']=='DESEMBOLSO A CTA CLIENTE']
if(len(debt_disbursement)<=0):
    debt_p_amount = 0
else:
    debt_p_amount = debt_disbursement['SALDO'].iloc[0]
debt_disbursement
```

```python
p_extraordinary_payment=0
ia_extraordinary_payment=0

#extraordinary_payments=zero_description[np.flatnonzero(np.char.startswith(zero_description, ' PAGO ABONO EXTRAORDINARIO'))]
extraordinary_payments=zero_description[np.flatnonzero(np.char.startswith(zero_description, 'Abono Extraordinario'))]
if(len(extraordinary_payments)>0):
    print('Pagos extraordinarios')
    print('-----------------------')
    extraordinary_payments=boct_zero[boct_zero['DESCRIPTION'] == extraordinary_payments[0]].sort_values(by='FETCHA TRANS', ascending = True)
    extraordinary_payment_journals=np.array(extraordinary_payments['NO SEGUIM'])
    for extraordinary_journal in extraordinary_payment_journals:
        extraordinary_payment=boct_non_zero[boct_non_zero['NO SEGUIM'] == extraordinary_journal]
        p_extraordinary_payment=p_extraordinary_payment+abs(extraordinary_payment[extraordinary_payment['PRIORITY']=='P ']['CANTIDAD'].sum())
        try:
            ia_extraordinary_payment=ia_extraordinary_payment+abs(extraordinary_payment[extraordinary_payment['PRIORITY']=='IA']['CANTIDAD'].sum())
        except:
            pass
        details=pd.DataFrame({"Tipo": extraordinary_payment['PRIORITY'],
                              "Monto": extraordinary_payment['CANTIDAD']})
        extraordinary_payment=extraordinary_payment[extraordinary_payment['PRIORITY'] == '  ']
        extraordinary_payment=np.array(extraordinary_payment[['FETCHA TRANS', 'CANTIDAD']])
        extraordinary_payment=extraordinary_payment[0]
        print(details)
        print(f'\n${abs(extraordinary_payment[1])} fueron acreditados el {extraordinary_payment[0].date()}')
        print('-----------------------')
    p_extraordinary_payment=round(p_extraordinary_payment,2)
    ia_extraordinary_payment=round(ia_extraordinary_payment,2)
    print(f'Total para el capital: ${p_extraordinary_payment}')
    print(f'Total para el interés: ${ia_extraordinary_payment}')
else:
    print('El cliente no tuvo pagos extraordinarios')
```

```python
p_reversed_payment=0
ia_reversed_payment=0
in_reversed_payment=0
mg_reversed_payment=0

reversed_payments=zero_description[np.flatnonzero(np.char.startswith(zero_description, 'REV.PAGO PRESTAMO'))]
if(len(reversed_payments)>0):
    print('Reversos')
    print('-----------------------')
    reversed_payments=boct_zero[boct_zero['DESCRIPTION'] == reversed_payments[0]].sort_values(by='FETCHA TRANS', ascending = True)
    reversed_payment_journals=np.array(reversed_payments['NO SEGUIM'])
    for reversed_journal in reversed_payment_journals:
        reversed_payment=boct_non_zero[boct_non_zero['NO SEGUIM'] == reversed_journal]
        reversed_payment=reversed_payment[reversed_payment['CANTIDAD']>=0]
        try:
            p_reversed_payment+=abs(reversed_payment[reversed_payment['PRIORITY']=='P ']['CANTIDAD'].iloc[0])
        except:
            pass
        try:
            ia_reversed_payment+=abs(reversed_payment[reversed_payment['PRIORITY']=='IA']['CANTIDAD'].iloc[0])
        except:
            pass
        try:
            in_reversed_payment+=abs(reversed_payment[reversed_payment['PRIORITY']=='IN']['CANTIDAD'].iloc[0])
        except:
            pass
        try:
            mg_reversed_payment+=abs(reversed_payment[reversed_payment['PRIORITY']=='MG']['CANTIDAD'].iloc[0])
        except:
            pass
        details=pd.DataFrame({"Tipo": reversed_payment['PRIORITY'],
                              "Monto": reversed_payment['CANTIDAD']})
        reversed_payment=reversed_payment[reversed_payment['PRIORITY'] == '  ']
        reversed_payment=np.array(reversed_payment[['FETCHA TRANS', 'CANTIDAD']])
        reversed_payment=reversed_payment[0]
        print(details)
        print(f'\n${abs(reversed_payment[1])} fueron reversados el {reversed_payment[0].date()}')
        print('-----------------------')
else:
    print('El cliente no tuvo reversos')
```

```python
boct_lower_than_zero=boct_non_zero[boct_non_zero['CANTIDAD'] < 0]
journals=np.array(boct_lower_than_zero['NO SEGUIM'])
journals=np.unique(journals)
movements=[]
error_journals=[]
error_amounts=[]
for journal in journals:
    movements.append(boct_lower_than_zero.groupby(['NO SEGUIM']).get_group(journal))
error_flag=False
unacredited_total=0
for i in range(len(movements)):
    try:
        movement=movements[i].groupby(['PRIORITY']).get_group('  ')
    except:
        continue
    total_accredited=round(float(sum(movements[i]['CANTIDAD'])-movement['CANTIDAD']),2)
    total_paid=round(float(movement['CANTIDAD']),2)
    if(total_paid!=total_accredited):
        error_flag=True
        unacredited=round(total_accredited-total_paid,2)
        unacredited_total+=unacredited
        details=pd.DataFrame({"Tipo": movements[i]['PRIORITY'],
                              "Monto": movements[i]['CANTIDAD']})
        print(details)
        print(f"\n${unacredited} no fueron acreditados en {int(np.mean(movements[i]['NO SEGUIM']))} ({movements[i]['FETCHA TRANS'].iloc[0].date()})")
        print('-----------------------')
        error_journals.append(np.mean(movements[i]['NO SEGUIM']))
        error_amounts.append(unacredited)
if(not error_flag):
    print('Todos los valores fueron acreditados')
else:
    print(f'Total: ${unacredited_total}')
```
